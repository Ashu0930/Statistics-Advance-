#1. > What is a random variable in probability theory?

-> In probability theory, a random variable is a variable that represents the possible outcomes of a random experiment in numerical terms.

There are two main types:
- Discrete Random Variable

Takes on countable values (e.g., 0, 1, 2, 3, …).

Example: The number of heads in 3 coin tosses.

- Continuous Random Variable

Takes on uncountably infinite values within an interval (e.g., any real number between 0 and 1).

Example: The time it takes for a train to arrive.

#2.  Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import random
import matplotlib.pyplot as plt

# Define possible outcomes (e.g., 1 to 6 like a die)
outcomes = [1, 2, 3, 4, 5, 6]

# Simulate frequencies
counts = {x: 0 for x in outcomes}
num_trials = 10000

for _ in range(num_trials):
    roll = random.choice(outcomes)
    counts[roll] += 1

# Convert counts to probabilities (PMF)
pmf = [counts[x] / num_trials for x in outcomes]

# Plotting
plt.bar(outcomes, pmf)
plt.title("PMF of Discrete Uniform Distribution")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.ylim(0, 0.25)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): The probability of success (0 ≤ p ≤ 1)

    Returns:
    float: Probability of the given outcome
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not 0 <= p <= 1:
        raise ValueError("Probability p must be between 0 and 1.")

    return p if x == 1 else 1 - p

print("P(X=0):", bernoulli_pdf(0, 0.7))  # Output: 0.3
print("P(X=1):", bernoulli_pdf(1, 0.7))  # Output: 0.7
```

#9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores for the given dataset and plots them.

    Parameters:
    data (list or np.array): Input numeric data
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='royalblue')
    plt.axhline(0, color='black', linestyle='--', linewidth=1)
    plt.title("Z-Scores of the Dataset")
    plt.xlabel("Index")
    plt.ylabel("Z-Score")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    return z_scores

# Example usage
data_sample = [10, 12, 15, 14, 8, 13, 11, 9, 16, 12]
z_scores = calculate_and_plot_z_scores(data_sample)
print("Z-scores:", z_scores)
```

#14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    z = (x - mean) / std
    return z

def plot_z_score(x, mean, std):
    # Calculate Z-score
    z = calculate_z_score(x, mean, std)
    print(f"Z-score of {x} (mean={mean}, std={std}): {z:.2f}")

    # Generate x-values for standard normal distribution
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals, 0, 1)  # Standard normal PDF

    # Plot the standard normal curve
    plt.plot(x_vals, y_vals, label='Standard Normal Distribution', color='blue')

    # Shade area under the curve up to the Z-score
    x_fill = np.linspace(-4, z, 1000)
    y_fill = norm.pdf(x_fill)
    plt.fill_between(x_fill, y_fill, alpha=0.4, color='skyblue', label=f'P(Z < {z:.2f})')

    # Mark the Z-score point
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title("Z-score on Standard Normal Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

# Example usage:
data_point = 72
mean = 70
std_dev = 4

plot_z_score(data_point, mean, std_dev)
```

#15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Critical z value for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)

    # p-value for two-tailed test
    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-score = {z:.4f}")
    print(f"Z-critical = ±{z_critical:.4f}")
    print(f"P-value = {p_value:.4f}")

    # Conclusion
    if abs(z) > z_critical:
        print("Reject the null hypothesis (statistically significant).")
    else:
        print("Fail to reject the null hypothesis (not statistically significant).")

# Example data
sample = [72, 68, 71, 69, 70, 73, 74, 70, 68, 72]
pop_mean = 70
pop_std = 3  # Population standard deviation assumed known

z_test(sample, pop_mean, pop_std)
```

#16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy import stats

# Sample data
data = [72, 68, 71, 69, 70, 73, 74, 70, 68, 72]
confidence = 0.95

# Sample statistics
n = len(data)
mean = np.mean(data)
std_err = stats.sem(data)  # Standard error of the mean

# T critical value for 95% CI and df = n - 1
t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

# Margin of error
margin = t_critical * std_err

# Confidence Interval
ci_lower = mean - margin
ci_upper = mean + margin

print(f"Sample Mean = {mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

#17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy import stats

# Step 1: Generate random data from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=100, scale=15, size=50)  # mean=100, std=15, n=50

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
std_err = stats.sem(data)  # Standard Error of the Mean

# Step 3: Calculate the 95% confidence interval using t-distribution
confidence = 0.95
n = len(data)
t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
margin_of_error = t_critical * std_err

# Step 4: Compute CI bounds
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 5: Output the result
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_val = 4  # average rate (λ)
x = 6           # value to evaluate the CDF at

# Calculate the CDF at x
cdf_value = poisson.cdf(x, mu=lambda_val)
print(f"P(X ≤ {x}) for Poisson(λ={lambda_val}) = {cdf_value:.4f}")

# Optional: Plot the CDF for x = 0 to 15
x_vals = np.arange(0, 16)
cdf_vals = poisson.cdf(x_vals, mu=lambda_val)

plt.step(x_vals, cdf_vals, where='post', color='darkgreen', label='Poisson CDF')
plt.scatter(x, cdf_value, color='red', zorder=5, label=f'P(X ≤ {x}) = {cdf_value:.2f}')
plt.title(f"Cumulative Distribution Function (CDF) of Poisson(λ={lambda_val})")
plt.xlabel("x")
plt.ylabel("P(X ≤ x)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

#21. Write a Python program to compare the standard deviations of two datasets and visualize the difference.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate two datasets
np.random.seed(0)
data1 = np.random.normal(loc=50, scale=5, size=100)   # Mean = 50, std = 5
data2 = np.random.normal(loc=50, scale=10, size=100)  # Mean = 50, std = 10

# Calculate standard deviations
std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

# Bar chart to compare standard deviations
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(['Dataset 1', 'Dataset 2'], [std1, std2], color=['blue', 'green'])
plt.title("Comparison of Standard Deviations")
plt.ylabel("Standard Deviation")
plt.grid(True, axis='y', linestyle='--', alpha=0.6)

# Histograms of both datasets
plt.subplot(1, 2, 2)
plt.hist(data1, bins=20, alpha=0.6, label='Dataset 1 (σ≈5)', color='blue')
plt.hist(data2, bins=20, alpha=0.6, label='Dataset 2 (σ≈10)', color='green')
plt.title("Distribution Comparison")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

#22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

```python
import numpy as np

# Step 1: Generate a dataset from a normal distribution
np.random.seed(1)
data = np.random.normal(loc=100, scale=15, size=100)

# Step 2: Calculate Range
data_range = np.max(data) - np.min(data)

# Step 3: Calculate IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Step 4: Display results
print(f"Range: {data_range:.2f}")
print(f"IQR (Interquartile Range): {iqr:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
```

#23. Implement Z-score normalization on a dataset and visualize its transformation.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a dataset
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=100)

# Step 2: Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Step 3: Print summary
print(f"Original Mean: {mean:.2f}, Original Std Dev: {std:.2f}")
print(f"Z-score Mean: {np.mean(z_scores):.2f}, Z-score Std Dev: {np.std(z_scores):.2f}")

# Step 4: Visualize original vs normalized
plt.figure(figsize=(12, 5))

# Original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title("Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.legend()

# Z-score normalized data
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=20, color='lightgreen', edgecolor='black')
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='--', label='Mean = 0')
plt.legend()

plt.tight_layout()
plt.show()
```

#24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution(size=1000, mean=0, std=1):
    # Generate data from a normal distribution
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    skew_val = skew(data)
    kurt_val = kurtosis(data)  # Fisher's definition (normal => 0)

    print(f"Skewness: {skew_val:.4f}")
    print(f"Kurtosis (Fisher): {kurt_val:.4f}")

    # Interpretation
    print("\nInterpretation:")
    if abs(skew_val) < 0.5:
        print("- Distribution is approximately symmetric.")
    elif skew_val > 0:
        print("- Distribution is right-skewed.")
    else:
        print("- Distribution is left-skewed.")

    if abs(kurt_val) < 0.5:
        print("- Kurtosis is near normal (mesokurtic).")
    elif kurt_val > 0:
        print("- Distribution has heavier tails (leptokurtic).")
    else:
        print("- Distribution has lighter tails (platykurtic).")

# Example usage
analyze_distribution()
```
